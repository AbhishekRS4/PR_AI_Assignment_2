```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from visualization_utils import plot_compare_scores
```

```python
dir_bovw_data = "bovw_features_data/"
```

```python
file_cv_gnb = os.path.join(dir_bovw_data, "image_classification_cv_gaussian_nb.csv")
file_test_gnb = os.path.join(dir_bovw_data, "image_classification_test_gaussian_nb.csv")

df_cv_gnb = pd.read_csv(file_cv_gnb)
df_test_gnb = pd.read_csv(file_test_gnb)
```

## Gaussian Naive Bayes - accuracy score comparison plot

```python
plot_compare_scores(df_cv_gnb.num_visual_words.to_numpy(), 
                    df_cv_gnb.best_cv_score.to_numpy(),
                    df_test_gnb.test_acc.to_numpy(),
                    str_title="Comparison of accuracy score cross validation vs test for Gaussian Naive Bayes")
```

## KNN - Accuracy score comparison plots

```python
file_cv_knn = os.path.join(dir_bovw_data, "image_classification_gs_cv_knn.csv")
file_test_knn = os.path.join(dir_bovw_data, "image_classification_test_knn.csv")

df_cv_knn = pd.read_csv(file_cv_knn)
df_test_knn = pd.read_csv(file_test_knn)

df_cv_knn_std_scaler = df_cv_knn[df_cv_knn["preprocess"] == "std_scaler"]
df_cv_knn_tf_idf = df_cv_knn[df_cv_knn["preprocess"] == "tf_idf"]
df_test_knn_std_scaler = df_test_knn[df_test_knn["preprocess"] == "std_scaler"]
df_test_knn_tf_idf = df_test_knn[df_test_knn["preprocess"] == "tf_idf"]
```

```python
plot_compare_scores(df_cv_knn_std_scaler.num_visual_words.to_numpy(), 
                    df_cv_knn_std_scaler.best_cv_score.to_numpy(),
                    df_cv_knn_tf_idf.best_cv_score.to_numpy(),
                    str_title="Comaprison of cross validation accuracy score for std_scaler vs tf_idf for KNN",
                    what_comparison="preprocess",
                    str_label_plot_1="best cross validation accuracy score - std_scaler",
                    str_label_plot_2="best cross validation accuracy score - tf_idf")
```

```python
plot_compare_scores(df_test_knn_std_scaler.num_visual_words.to_numpy(), 
                    df_test_knn_std_scaler.test_acc.to_numpy(),
                    df_test_knn_tf_idf.test_acc.to_numpy(),
                    str_title="Comaprison of test accuracy score for std_scaler vs tf_idf for KNN",
                    what_comparison="preprocess",
                    str_label_plot_1="best test accuracy score - std_scaler",
                    str_label_plot_2="best test accuracy score - tf_idf")
```

```python
plot_compare_scores(df_cv_knn_std_scaler.num_visual_words.to_numpy(), 
                    df_cv_knn_std_scaler.best_cv_score.to_numpy(),
                    df_test_knn_std_scaler.test_acc.to_numpy(),
                    str_title="Comaprison of accuracy score with std_scaler for cross validation vs test for KNN")
```

```python
plot_compare_scores(df_cv_knn_tf_idf.num_visual_words.to_numpy(), 
                    df_cv_knn_tf_idf.best_cv_score.to_numpy(),
                    df_test_knn_tf_idf.test_acc.to_numpy(),
                    str_title="Comaprison of accuracy score with tf_idf for cross validation vs test for KNN")
```
