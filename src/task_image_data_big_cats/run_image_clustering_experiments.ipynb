## 1) Find the optimal range for eps to be used for DBSCAN clustering

```python
from learning_utils import get_nearest_neighbors, get_tfidf_transformer, get_learning_pipeline
```

```python
import matplotlib.pyplot as plt
```

```python
dir_sift_data = "sift_features/"
```

```python
num_visual_words = 100
```

```python
train_x = np.load(os.path.join(dir_sift_data, f"train_sift_{num_visual_words}.npy"))
test_x = np.load(os.path.join(dir_sift_data, f"test_sift_{num_visual_words}.npy"))
```

```python
all_train_data = np.concatenate((train_x, test_x), axis=0)
```

```python
preprocess = "tf_idf"
preprocessor = get_tfidf_transformer()
nn_algo = get_nearest_neighbors()
pipeline = [(preprocess, preprocessor), ("nn", nn_algo)]
```

```python
nn_pipeline = get_learning_pipeline(pipeline)
```

```python
nn_pipeline.fit(all_train_data)
```

```python
distances, indices = nn_pipeline["nn"].kneighbors(nn_pipeline["tf_idf"].transform(all_train_data))
```

```python
distances = np.sort(distances, axis=0)
```

```python
distances.shape
```

```python
fig = plt.figure(figsize=(8, 5))
plt.title(f"Num visual words : {num_visual_words}, k-distance elbow plot for finding optimum range of eps")
plt.ylabel("distance")
plt.xlabel("samples")
plt.grid()
plt.plot(distances[:, 1])
plt.show()
```
